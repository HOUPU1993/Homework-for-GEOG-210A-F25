<font color="orange">a).f(x) = e^x - x - 3</font>

```python
# I set up the initial_guess as 5 and -5 seperately
def f(x):
    return np.exp(x) - x - 3

def newton_iter(initial_guess):
    x = initial_guess
    y = f(x)
    xi, yi = [x], [y]
    dx = 0.001
    itc = -1  

    while abs(y) > 0.0001:
        yplus = f(x + dx)
        yminus = f(x - dx)
        yprime = (yplus - yminus) / (2 * dx)
        x = x - y / yprime
        y = f(x)
        xi.append(x)
        yi.append(y)
        itc += 1
        if len(xi) > 100:  
            break
    

    print('the root of the function near x=', initial_guess, 'is' ,x, ',and the iterations is', itc)
    return xi, yi

xi1, yi1 = newton_iter(-5)
xi2, yi2 = newton_iter(5)

x_vals = np.linspace(-5, 5, 100)
plt.plot(x_vals, f(x_vals), label="f(x) = e^x - x - 3", color='gray', alpha=0.5)

plt.plot(xi1, yi1, marker='o', label='initial = -5', color='C1')
plt.plot(xi2, yi2, marker='s', label='initial = 5', color='C2')

plt.axhline(0, color='black', lw=1)
plt.grid(True, linestyle=":")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Newton Iteration from Two Initial Guesses")
plt.show()
```
