<font color="orange">a).f(x) = e^x - x - 3</font>

```python
# I set up the initial_guess as 5 and -5 seperately
def f(x):
    return np.exp(x) - x - 3

def newton_iter(initial_guess):
    x = initial_guess
    y = f(x)
    xi, yi = [x], [y]
    dx = 0.001
    itc = -1  

    while abs(y) > 0.0001:
        yplus = f(x + dx)
        yminus = f(x - dx)
        yprime = (yplus - yminus) / (2 * dx)
        x = x - y / yprime
        y = f(x)
        xi.append(x)
        yi.append(y)
        itc += 1
        if len(xi) > 100:  
            break
    

    print('the root of the function near x=', initial_guess, 'is' ,x, ',and the iterations is', itc)
    return xi, yi

xi1, yi1 = newton_iter(-5)
xi2, yi2 = newton_iter(5)

x_vals = np.linspace(-5, 5, 100)
plt.plot(x_vals, f(x_vals), label="f(x) = e^x - x - 3", color='gray', alpha=0.5)

plt.plot(xi1, yi1, marker='o', label='initial = -5', color='C1')
plt.plot(xi2, yi2, marker='s', label='initial = 5', color='C2')

plt.axhline(0, color='black', lw=1)
plt.grid(True, linestyle=":")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Newton Iteration from Two Initial Guesses")
plt.show()
```

<font color="orange">b)
Assuming  z_0 = 100m, find the value of \( b \) that minimizes the misfit (in the least-squares sense) between the flux predicted by a power-law model of the form  

$
F = F_{z_0} \left( \frac{z}{z_0} \right)^{-b}
$

and the observed flux from the sediment traps.Use the procedure outlined in class to set up a **cost function** that measures the model–data misfitand determine the minimum of the cost function using **Newton’s method** on the cost function derivative.
Write your own Python scripts/functions for this — **do not use canned functions.**  Use a starting guess of \( b = 0.1 \).</font>

```python
sediment_trap_data
```

```python
z = sediment_trap_data['depth'].values
Fobs = sediment_trap_data['organic_carbon_flux'].values

# reference depth
z0 = 100
# observed flux at z0 (or you can take the first value directly)
Fz0 = Fobs[0]


# --- define model function and cost function ---
def model(b, z, Fobs):
    """Compute modeled flux and total squared misfit (J(b))"""
    Fmod = Fz0 * (z / z0) ** (-b)  # power-law model
    J = np.sum((Fmod - Fobs) ** 2)  # sum of squared differences
    return Fmod, J


# set up an iteration loop for Newton's method
b = 0.1  # initial guess
dJdb = 1  # initial guess

# --- Newton's method iteration ---
while abs(dJdb) > 1e-3: # use a .001 tolerance to find the zero of the derivative
  # "run the model" and determine the cost
  Fmod, J = model(b, z, Fobs)
  # compute J'(b)
  db = 1e-3  # finite difference
  Fup, Jup = model(b + db, z, Fobs)      # forward evaluation
  Fdown, Jdown = model(b - db, z, Fobs)  # backward evaluation
  dJupdb = (Jup - J) / db
  dJdowndb = (J - Jdown) / db
  dJdb = 0.5 * (dJupdb + dJdowndb)       # centered difference
  # compute J''(b)
  d2Jdb2 = (dJupdb - dJdowndb) / db      # second derivative
  # Newton step
  nstep = -dJdb / d2Jdb2                 # Newton step = -f(x)/f'(x)
  # new value of b
  b = b + nstep
# --- the zero (optimum b) ---
print('The optimal value of b is', b)
```

```python
z = sediment_trap_data['depth'].values
Fobs = sediment_trap_data['organic_carbon_flux'].values

# reference depth
z0 = 100
# observed flux at z0 (or you can take the first value directly)
Fz0 = Fobs[0]


# --- define model function and cost function ---
def model(b, z, Fobs):
    """Compute modeled flux and total squared misfit (J(b))"""
    Fmod = Fz0 * (z / z0) ** (-b)  # power-law model
    J = np.sum((Fmod - Fobs) ** 2)  # sum of squared differences
    return Fmod, J


# set up an iteration loop for Newton's method
b = 0.1  # initial guess
dJdb = 1  # initial guess

b_record = []
J_record = []
dJ_record = []


# --- Newton's method iteration ---
while abs(dJdb) > 1e-3: # use a .001 tolerance to find the zero of the derivative
  # "run the model" and determine the cost
  Fmod, J = model(b, z, Fobs)
  # compute J'(b)
  db = 1e-3  # finite difference
  Fup, Jup = model(b + db, z, Fobs)      # forward evaluation
  Fdown, Jdown = model(b - db, z, Fobs)  # backward evaluation
  dJupdb = (Jup - J) / db
  dJdowndb = (J - Jdown) / db
  dJdb = 0.5 * (dJupdb + dJdowndb)       # centered difference
  # compute J''(b)
  d2Jdb2 = (dJupdb - dJdowndb) / db      # second derivative
  # Newton step
  nstep = -dJdb / d2Jdb2                 # Newton step = -f(x)/f'(x)
  # new value of b
  b = b + nstep
  
  b_record.append(b)
  J_record.append(J)
  dJ_record.append(dJdb)

print(f"The optimal value of b is {b:.6f}, Iterations = {len(b_record)}")

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 8), sharex=True)

ax1.plot(b_record, J_record, 'o-', color='teal')
ax1.set_ylabel('Cost Function J(b)')
ax1.set_title('Newton Iteration: b vs Cost Function')
ax1.grid(True, linestyle=':')

ax2.plot(b_record, dJ_record, 'o-', color='orange')
ax2.set_xlabel('b value')
ax2.set_ylabel("Derivative J'(b)")
ax2.set_title('b vs Derivative of Cost Function')
ax2.grid(True, linestyle=':')

plt.tight_layout()
plt.show()
```
